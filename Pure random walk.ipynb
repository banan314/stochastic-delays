```python
import matplotlib.pyplot as plt
import random
import numpy as np
import math
from delay.delay import *
```

```python
def simulate():
    def nextElem(func):
        term = -N/2 if func == max else N/2
        value = -1 if func == max else 1
        return func(x[current] + value, term)
    
    current = len(x)-1
    for _ in range(duration):
        prob = 1/2
        r = random.random()
        if r < prob:
            nextElement = nextElem(min) # +1
        else:
            nextElement = nextElem(max) # -1
        
        x.append(nextElement)
        current += 1
```

```python
N = 500
x = [0] * 20 + [1] 
duration = 10000 * N

simulate()

_, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 4.8))
ax1.plot(x)
(bins, _, _) = ax2.hist(x, bins=N+1)
print('Âµ = ', np.mean(x))
```
